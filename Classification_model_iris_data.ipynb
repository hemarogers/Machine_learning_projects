# 1. Import Libraries

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
import pandas as pd
```

# 2. Load iris data, input featurs -feature names, output-target

```python
iris=load_iris()
type(iris)
# variable information
#print(iris.data)
#print(iris.feature_names)
#print(iris.target)
#print(iris.target_names)
#print(iris.target)
df=pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target']=iris.target
df['species']=iris.target_names[iris.target]
df.head(59)
```

# 3. Assign the input and output variables

```python
X=iris.data
y=iris.target
```

# 4. Build a classification model

```python
#Build the classification model with all the data and do some initial predictions
clf=RandomForestClassifier()
clf.fit(X,y)
```

```python
#Feature importance
print(clf.feature_importances_)
```

```python
#look at the first row, do the prediction for first row it gives the target value
X[0]
print(clf.predict([[5.1, 3.5, 1.4, 0.2]]))
print(clf.predict_proba(X[[0]])) #shows that the first row is label 1- setosa
```

# 5. Data split-80% for the training and 20% for testing

```python
#X is the feature, y is the target
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.2)
X_test.shape,y_test.shape
```

```python
X_train.shape, y_train.shape
```

# 6. Rebuild random forrest

```python
clf.fit(X_train,y_train)
```

# 7. Perform prediction on a single row

```python
clf.predict(X[[0]])
```

# 8. Predict for X_test and compare it with y_test -X_test is the feature, y_test is the target

```python
# predict for X_test
print(clf.predict(X_test))
# compare with y_test
print(y_test)
```

# 9. Model performance

```python
print(clf.score(X_test,y_test))
```
